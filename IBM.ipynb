```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from joblib import dump, load
```

```python
data = pd.read_csv('heart.csv')
```

```python
imputer = SimpleImputer(strategy='mean')
data.iloc[:, :-1] = imputer.fit_transform(data.iloc[:, :-1])
```

```python
data = data.drop_duplicates()
```

```python
X = data.drop('target', axis=1)
y = data['target']
```

```python
categorical_features = ['sex', 'cp', 'restecg', 'slope', 'ca', 'thal']
one_hot_encoder = OneHotEncoder()
X_encoded = one_hot_encoder.fit_transform(X[categorical_features]).toarray()
X = X.drop(categorical_features, axis=1)
X = np.concatenate([X.values, X_encoded], axis=1)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
```

```python
new_data_prediction = logistic_model.predict([X_test[0]])
```

```python
dump(logistic_model, 'logistic_model.joblib')
dump(svc_model, 'svc_model.joblib')
dump(knn_model, 'knn_model.joblib')
dump(dt_model, 'dt_model.joblib')
dump(rf_model, 'rf_model.joblib')
dump(gb_model, 'gb_model.joblib')
```
